зависимость региона от типа продаж(онлайн-оффлайн)


```python
y1 = df[df['sales channel'] == 'Offline']
y2 = df[df['sales channel'] == 'Online']

plt.figure(figsize=(12,6))
plt.hist(y2['region'], color='blue',alpha=0.2, label='online', bins=30)
plt.hist(y1['region'], color='red',alpha=0.2, label='offline', bins=30)
plt.xlabel('Region')
plt.ylabel('Sales Channel')
plt.xticks(rotation=45)

plt.legend()
plt.show()
```

разделяем таблицу по item type


```python
items = df['item type'].unique()
for i in range(len(items)): print(f"{i+1}.{items[i]}")
```

```python
food = ['Baby Food', 'Cereal','Meat','Snacks','Beverages','Meat','Fruits','Vegetables']
df['food or not_food'] = df['item type']
for columns, series in df.items():
    if columns == 'food or not_food':
        for i in range(len(series)):
            if series[i] in food: series[i] = 'food'
            else: series[i] = 'things'
df
```

Построим гистограмму по доходу от продаж по типам товаров в зависмости от типа
продажи (2 графика)

один график(гистограмма) зависимость типа товара из food по доходам
второй график(гистограмма) зависимость типа товара из things также по доходам

```python
group = df.groupby(['food or not_food','item type'])['total revenue'].sum()
group
```

```python
group = df.groupby(['food or not_food','item type'])['total revenue'].sum().plot(ylabel='total revenue', xlabel='food or not_food',kind='bar', color='purple')
group
```

Просуммируем данные по Total Profit по годам и построим гисторгамму по региону в
